```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_image(img, grayscale=True):
    plt.axis('off')
    if grayscale:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv2.imread("discovery.jpg")
print(img.shape)

# Para realizar un reescalamiento
# De esta forma nosotros establecemos las dimensiones
# img2 = cv2.resize(img,(500,350)) 
# Nosotros establecemos una relacion sobre la imagen principal
img2 = cv2.resize(img,None,fx=0.5,fy=0.5)
print(img2.shape)
plot_image(img2,False)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i in range(0,3):
    axes[i].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)[:,:,i])
    # axes[i].set_title('Imagen Original')
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```

```python
#Para aplicar en escala de grises
gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
# Se eliminan las 3 capas siendo ahora una sola correspodiente a la escala de grises
print(gray.shape)
plot_image(gray)
```
